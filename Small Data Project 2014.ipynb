Ultimately, my biggest question is whether the project was a success: did I do at least three things more than I didn't?

To figure that out, I calculate the cumulative sum of activities done vs. laziness. I felt like it was necessary to reweight my laziness. if I didn't do something during the day, I wasn't 3x more lazy than when I wasn't doing something.

```python
activities = np.cumsum(df.active_sum)
activities_diff = np.cumsum(df.active_sum.diff())
laziness = np.cumsum(df.D) / 3
alf = activities/laziness
dlf = activities_diff/laziness

alf.plot()
plt.xlabel('date')
plt.ylabel('activity : laziness ratio')
plt.show()
```

It seems I was most productive at the middle of the project, steadily climbing from  April into July. That's probably because school was out and we hadn't started traveling yet. Here's a difference version of the same graph, with January dropped because of the extreme variance:

```python
dlf[30:].plot()
plt.xlabel('date')
plt.ylabel('activity : laziness ratio')
plt.show()
```

Regression toward the mean much? There was a lot of variance at the beginning of the year and then the ratio of activities to laziness leveled off significantly. Taking into account scaling, I did a pretty good job of getting things done, but I would have liked to be more consistently productive toward the end of the year.
