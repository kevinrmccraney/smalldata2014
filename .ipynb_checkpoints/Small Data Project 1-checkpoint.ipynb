```python
# Hey. This is the import section.

%matplotlib inline
from matplotlib import pyplot as plt
plt.style.use('fivethirtyeight')

from scipy import stats
from scipy.stats.kde import gaussian_kde

import seaborn as sns
sns.set_palette("Paired", desat=.9)

import numpy as np
import pandas as pd

import csv
import collections
```

```python
# Originally, I took my notes in a rtf file with a little bit of additional information. How foolish.
# The first step was cleanup. I got rid of newline characters, and so on and so on.
# The goal here is to make everything organized into an almost-CSV-ready thing.
# Excuse the mess. It's largely irrelevant to the results.

f = open('smalldata.txt', 'r')
sd = f.read().splitlines()

day_counts = {sd[i][0:5]+'/2014': collections.Counter(''.join(sorted(sd[i][8:]))) 
              for i in xrange(365)}

concat_strings = {sd[i][0:5]+'/2014': ''.join(sorted(sd[i][8:])) for i in xrange(365)}

# I copied these into notepad, did some regex and tedious reformatting.
# It was not a good time. Thankfully I had the foresight to make sure the text was all the same case.
```

```python
# Here goes something. I imported the CSVs and joined them together.

df1 = pd.read_csv('dates.csv', parse_dates=["Dates"], index_col='Dates')
df2 = pd.read_csv('strings.csv', parse_dates=["Dates"], index_col='Dates')

df = df1.join(df2)

df.head()
```

```python
# Now, let's see what we got.

df.plot()
plt.xlabel('date')
plt.ylabel('activity level')
plt.show()
```

```python
# That looks horrendous and is indecipherable. Let's try weekly
# Let's resample by week to make it easier on the eyes.

weeks = df.resample('W', kind='period')

weeks.plot()
plt.xlabel('date')
plt.ylabel('activity level')
plt.show()

# This is okay. We get a lot of nice variance going on.
```

```python
# Let's try months though, just to check it out.

months = df.resample('M', kind='period')

months.plot()
plt.xlabel('date')
plt.ylabel('activity level')
plt.show()
```

```python
# and just for fun, here's quarterly. 
# It seems quarter 3 was great for not doing anything last year, while quarter 2 was very productive.

quarters = df.resample('Q', kind='period')

quarters.plot()
plt.xlabel('date')
plt.ylabel('activity level')
plt.show()
```

```python
# Let's start with an easy one. We don't even need the graph. What specific day did I do the most stuff?

df['active_sum'] = df['X'] + df['Y'] + df['Z'] + df['A'] + df['B'] 
print df.loc[df['active_sum'] == max(df.active_sum)]

# And a weirder one... how many things did I do in the 80th percentile of my productive days?
# I already know the bottom 20% is going to be composed of nothing, 
# because I traveled a lot last year, so I won't even bother.

topeighty = df.loc[df['active_sum'] == np.percentile(df.active_sum, 80)]
topeighty[12:25]

# I'm really curious about April/May being so well represented in terms of stuff I did.
# It was raining a lot then, so I probably spent quite a bit of time inside. Apparently writing and doing music.
```

```python
# Ultimately, my biggest question is whether the project was a success:
# did I do at least three things more than I didn't?
#
# To figure that out, I calculate the cumulative sum of activities done vs. laziness.
# I felt like it was necessary to reweight my laziness. if I didn't do something during the day, I wasn't 3x
# more lazy than when I wasn't doing something.

activities = np.cumsum(df.active_sum)
activities_diff = np.cumsum(df.active_sum.diff())
laziness = np.cumsum(df.D) / 3
alf = activities/laziness
dlf = activities_diff/laziness

alf.plot()
plt.xlabel('date')
plt.ylabel('activity : laziness ratio')
plt.show()
```

```python
# It seems I was most productive at the middle of the project, 
# steadily climbing from  April into July. That's probably because school was out
# and we hadn't started traveling yet.
# Here's a difference version of the same graph.

dlf.plot()
plt.xlabel('date')
plt.ylabel('activity : laziness ratio')
plt.show()

# Regression toward the mean much? There was a lot of variance at the beginning of the year
# and then the ratio of activities to laziness leveled off significantly. Taking into account scaling,
# I did a pretty good job of getting things done, but I would have liked to be more consistently productive
# toward the end of the year.
```
